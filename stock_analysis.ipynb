First Stock Data Collect

```python
import yfinance as yf

# Apple stock data download
stock = yf.download("AAPL", start="2023-01-01", end="2025-01-01")

# First 5 rows show
stock.head()
```

Graph Creation

```python
import matplotlib.pyplot as plt

# Closing price graph
plt.figure(figsize=(10,5))

plt.plot(stock['Close'])

plt.title("Apple Stock Price")
plt.xlabel("Date")
plt.ylabel("Price")

plt.show()
```

Moving Average

```python
# 20-day moving average
stock['MA20'] = stock['Close'].rolling(20).mean()

# Graph
plt.figure(figsize=(10,5))

plt.plot(stock['Close'], label='Close Price')
plt.plot(stock['MA20'], label='20 Day Moving Average')

plt.title("Apple Stock Price with Moving Average")

plt.legend()

plt.show()
```

Simple Prediction Model Building

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Create day numbers
stock = stock.dropna()

stock['Days'] = np.arange(len(stock))

# X and y values
X = stock[['Days']]
y = stock['Close']

# Model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict values
stock['Predicted'] = model.predict(X)

# Graph
plt.figure(figsize=(10,5))

plt.plot(stock['Close'], label='Actual Price')
plt.plot(stock['Predicted'], label='Predicted Trend')

plt.legend()
plt.title("Stock Price Prediction")

plt.show()
```

Saved Dataset as CSV

```python
stock.to_csv("apple_stock_data.csv")
```
